```python
# Puzzle #3
# This function is supposed to count the frequency of words in a sentence.
# But it’s buggy and sometimes crashes or gives wrong results.
# Correct the code to make it work properly.

def word_count(sentence: str) -> dict:
    counts = {}
    words = sentence.split(" ")
    for w in words:
        if w in counts:
            counts[w] = counts[w] + 1
        else:
            counts[w] += 1
    return counts

print(word_count("apple banana apple orange banana apple"))
# Expected: {'apple': 3, 'banana': 2, 'orange': 1}
```

```python
# Puzzle #6
# This function computes a cumulative sum of the numbers in a list.
# It returns a new list where each element is the sum of all previous elements plus the current one.
# Fix the off-by-one error to produce correct results.

def cumulative_sum(numbers):
    result = []
    for i in range(len(numbers)):
        if i == 0:
            result.append(numbers[i])
        else:
            result.append(result[i] + numbers[i])
    return result

print(cumulative_sum([1, 2, 3, 4]))
# Expected: [1, 3, 6, 10]
```
